```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('pistachio.csv')
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder #Change target class
```

```python
le = LabelEncoder()
```

```python
##df.loc[df['Class']== 'Siit_Pistachio']
```

```python
df['Class']=le.fit_transform(df['Class'])
```

```python
#0 - Kirmizi_Pistachio

#1 - Siit_Pistachio
```

```python
df
```

```python
X=df.drop(['Class'],axis=1) # X all columns except target
```

```python
y=df['Class'] #Y is target class
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X=scaler.fit_transform(X)
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lgr=LogisticRegression()
param_grid = {
    'penalty': ['l1','l2'],
    'solver' :['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky'],
    'max_iter': [1000,1500,2000],
    'multi_class' :['auto', 'ovr', 'multinomial'],
    'class_weight' :['dict','balanced']
}
```

```python
from sklearn.model_selection import GridSearchCV
lgr_grid=GridSearchCV(estimator=lgr,param_grid=param_grid,cv=3,verbose=2,n_jobs=4,return_train_score=False)
```

```python
lgr_grid.fit(X_train,y_train)
```

```python
lgr_grid.best_params_
```

```python
log=LogisticRegression(class_weight='balanced',max_iter=1000,multi_class='auto',penalty='l1',solver='liblinear')
```

```python
log.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred2=log.predict(X_test)
score_log=accuracy_score(y_test,y_pred2)
score_log
```

```python
import sklearn.metrics as metrics
from sklearn.metrics import classification_report
cm = metrics.confusion_matrix(y_test, y_pred2)
print(classification_report(y_test, y_pred2))
```
